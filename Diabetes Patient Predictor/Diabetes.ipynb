```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score


```

```python
#loading the dataset from kaggle
df = pd.read_csv('/content/diabetes.csv')
df
```

# **Data Collection and analysis**

```python
df.head()
```

```python
#to get the statistical measure of diabetic and non-diabetic
df['Outcome'].value_counts()
df.groupby('Outcome').mean()
```

```python
#we will seperate the outcomes and the other columns

X = df.drop(columns='Outcome',axis=1)
y = df['Outcome']
```

#**Data Standardization**

```python
#we did this because it will be easy for model to predict the data
scaler = StandardScaler()
scaler.fit(X)
```

```python
standardize_data = scaler.transform(X)
#again putting in X
X = standardize_data
```

# Data And Model Training

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,stratify=y,random_state=2)
```

```python
 #Training the model
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#Training the support vector Machine Classifier
classifier.fit(X_train,y_train)
```

# **Evaluation of Model**

```python
#accuracy score on the training data

X_train_pred = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred,y_train)
```

```python
print(f"accuracy score of the model is:{training_data_accuracy*100}%")
```

```python
#accuracy score on test data

X_train_pred = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_train_pred,y_test)

print(f"accuracy score of the model is:{test_data_accuracy*100}%")
```

```python
#we can see after both score that data model is not overfitted /overfitted
```

# **Making the predictive function**

```python
import pickle
with open('diabetes_model.pkl', 'wb') as file:
    pickle.dump(classifier, file)

# Save the scaler to a pickle file if you need to standardize inputs in the same way during inference
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```
